```julia
using Pkg
Pkg.activate("..")
using Revise
using PythonPlot,Statistics,Distributions,DataFrames,CSV,Random
PythonPlot.svg(true)
using SizeStructuredPopulations
using SpecialFunctions
include("./../paper-figures/examples.jl")

methods(generate_cell)

# seed the random number generator for reproducibility
Random.seed!(1234);
```

```julia

model = GrowthModel(M1()...)
init = vcat([0.0], [0.0, 0.0], zeros(3))
dt = 0.01
Tmax = 10.0

function terminate(cell)
    return cell.time[end] > Tmax 
end

cell = generate_cell(model, init,10.0,dt = 0.01);
@time grow_tree!(cell,model,terminate)
cell = generate_cell(model, init,10.0,dt = 0.01);
@time grow_tree!(cell,model,terminate,parallel_depth=6)
```

```julia
fig,ax = subplots(figsize=(4,4))

x0,y0,depth,hh,lw = 0.0,0.0,4,1,1
recursive_draw_tree(ax,cell,x0,y0,depth,hh,lw)
```
